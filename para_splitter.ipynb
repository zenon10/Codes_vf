##**Libraries**

```python
import pandas as pd
import re
```

#**Function**

```python
def para_splitter(input_text_file_path):

    f = open(input_text_file_path, "r", encoding="utf8")
    f=f.read()

    #splitting text file on double new lines
    paragraph = f.split('\n\n')

    #creating dataframe
    df = pd.DataFrame()

    para_id = []
    is_section_ind = []
    section_id = []
    is_article_ind = []
    article_id = []

#**********************************************************#
    for i in range(len(paragraph)):
        para_id.append(i)

    df['para_id'] = para_id
    df['para_text'] = paragraph
      
#**********************************************************#  
    for i in df['para_text']:
        if re.match('^[0-9][.][0-9][\s][a-zA-Z]',i):
            is_section_ind.append(1)
        else:
            is_section_ind.append(0)
        if re.match('^[Aa][rR][tT][iI][cC][lL][eE][.][0-9]',i):
            is_article_ind.append(1)
        else:
            is_article_ind.append(0)

    df['is_article_ind'] = is_article_ind
    df['is_section_ind'] = is_section_ind
#**********************************************************# 
    count = 1
    for i in df['is_section_ind']:
        if i == 1:
            section_id.append('s'+str(count))
            count = count +1
        else:
            section_id.append(None)

#**********************************************************#
    count = 1
    for i in df['is_article_ind']:
        if i == 1:
            article_id.append('a'+str(count))
            count = count +1
        else:
            article_id.append(None)

#**********************************************************#
    
    df['section_id'] = section_id
    df['article_id'] = article_id

    # applying ffill() method to fill the missing values
    df['section_id']=df['section_id'].ffill(axis = 0)
    df['article_id']=df['article_id'].ffill(axis = 0)

#**********************************************************#
    df = df[['para_id','para_text','is_article_ind','article_id','is_section_ind','section_id']]

    return df

    # df.to_excel('s04_cctp_montpellier v2_ocr2.xlsx', encoding="utf8" )
```

```python
input_text_file_path = r'/content/drive/MyDrive/Zenon_POC_OCR/s04_cctp_montpellier v2.txt'
df=para_splitter(input_text_file_path)
```
